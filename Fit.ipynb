```python
import math
import numpy as np
from numpy.linalg import inv

#Covariance matrix electrone mode
cov_e= np.array([(1.000000000,  0.49617583, -0.81046853,  0.31330784),
                 ( 0.49617583,  1.00000000, -0.65300820, -0.16417718),
                 (-0.81046853, -0.65300820,  1.00000000, -0.03983475),
                 ( 0.31330784, -0.16417718, -0.03983475,  1.00000000)])

#Covariance matrix muon mode
cov_m= np.array([(1.00000000,  0.55819379, -0.82065930,  0.30639805),
                 (0.55819379,  1.00000000, -0.72263757, -0.16258792),
                 (-0.8206593, -0.72263757,  1.00000000, -0.02897212),
                 (0.30639805, -0.16258792, -0.02897212,  1.00000000)])

#measurement result electrone mode
result_e=[1.165799794661480, 1.181991007093607, 0.8486768811375567, 0.03640308320620868]

#measurement result muon mode
result_m=[1.154766058936931, 1.291749191046493, 0.8474735659701261, 0.03545285029527105]

Cov_inv_e= inv(cov_e) #invert covariance matrix electrone mode
Cov_inv_m= inv(cov_m) #invert covariance matrix muon mode

#define fit-function
def Chi2_em(rho2, R1_1, R2_1, Vcb):
    print(rho2, R1_1, R2_1, Vcb)
    result=[rho2, R1_1, R2_1, Vcb]
    Chi21=0.
    for n in range(0,4):
        for m in range(0, 4):
            Chi21 = Chi21 + (result_e[n]-result[n])*Cov_inv_e[n,m]*(result_e[m]-result[m])
    for n in range(0,4):
        for m in range(0, 4):
            Chi21 = Chi21 + (result_m[n]-result[n])*Cov_inv_m[n,m]*(result_m[m]-result[m])
    return Chi21
```

```python
from iminuit import Minuit

Chi2_em.errordef = Minuit.LEAST_SQUARES

minu = Minuit(Chi2_em, rho2 =1.102, R1_1=1.158, R2_1=0.842, Vcb=0.03861)
minu.limits['rho2'] = (0.,2.)
minu.limits['R1_1'] = (0.,2.)
minu.limits['R2_1'] = (0.,2.)
minu.limits['Vcb'] = (0.,0.1)

minu.migrad()  # run optimiser

minu.hesse()   # run covariance estimator

np.savetxt("result_1em.txt", minu.values, delimiter=",", newline=',\n',)
np.savetxt("result_2em.txt", minu.errors, delimiter=",", newline=',\n',)
np.savetxt("result_3em.txt", minu.covariance, delimiter=",", newline=',\n',)

result_1em = minu.values
result_2em = minu.errors
result_3em = minu.covariance
print(result_1em)
print(result_2em)
print(result_3em)
```
